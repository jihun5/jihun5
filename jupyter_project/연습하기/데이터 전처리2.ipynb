# "많이 하는 질문!"
```iloc VS loc 뭐가 다른거죠 ?????```
- iloc는 인덱스와 컬럼을 리스트 배열로 선택하는 것!
- loc는 인덱스와 컬럼을 문자로 선택하는 것!
- 따라서 상황에 맞게 선택하는 것이 중요! 

````python
# 조건 필터링 걸어서 가져오기
# 30대 이상만 가져오기
df[df['age']>30]
````

````python
# job이 intern인 사람 가져오기
df[df['job']=='intern']
````

````python
# 조건 여러개 걸기 30대 이상 40대 이하
df[(df['age']>=30) & (df['age']<=40)] # & = and
````

````python
# 조건 여러개 걸기 30대 미만 혹은 40대 초과대 초과
df[(df['age']<30) | (df['age']>40)]
````

````python
# in을 통한 포함 조건 걸기(이렇게 하면 다음과 같은 에러가 발생)
df[df['job'] in ['student','manager']]
````

- DataFrame 그룹 생성

- pandas의 groupby() 연산자를 사용하여 집단, 그룹별로 데이터를 집계 및 요약을 할 수 있다.
- dataframe을 지정한 그룹으로 나누고, 각 그룹별로 집계합수를 적용하고 그룹별 집계 결과를 하나로 합치는 과정을 거친다.


````python
df = pd.read_csv('abalone.csv', names=['sex','length','diameter','height', 
                  'whole_weight','shucked_weight','viscera_weight', 
                  'shell_weight','rings'])
df.head()
````

````python
# 데이터 shape확인 -> 4177개의 데이터와 9개의 변수
df.shape
````

````python
# 데이터 결측값 확인
df.isnull().sum()
````

````python
# 기술통계 확인(평균, 편차, 최소값, 25%, 50%, 75%, 최대값)
df.describe()
````

````python
# 전복 성별(m,f,i)에 따라 groupby 함수를 통해 집계를 할 예정
# 그룹별 합계 : grouped.sum()
# 그룹별 평균 : grouped.mean()
# 그룹별 크기 : grouped.size()
# ex) DataFrame[집계 변수].groupby(DataFrame[집계 대상])
grouped = df['whole_weight'].groupby(df['sex'])

#grouped는 pandas의 SeiresGroupBy object 형태이다.
grouped
````

````python
grouped.sum()
````

````python
grouped.mean()
````

````python
grouped.size()
````

````python
# 그룹변수가 하나가 아닌, 전체 연속형 변수에 대한 집계
df.groupby(df['sex']).mean()
````

````python
# 다음과 같이 간단하게 작성할 수 있으며, 위와 같은 결과를 얻을 수 있다.
df.groupby('sex').mean()
````

````python
# 새로운 조건에 맞는 변수 추가
df['length_bool'] = np.where(df['length']>df['length'].median()# 중앙값 계산,
                             'length_long', # True일 경우
                             'length_short' # False 일경우)
                            )
df['length_bool']
````

````python
# 그룹변수를 2개 이상 선택하여 총계정리
df.groupby(['sex', 'length_bool']).mean()
````

````python
# 더 간결하게 표현한다면
## 그룹변수('sex', 'length_bool') 선택 및 집계변수('whole-weight') 평균
df.groupby(['sex', 'length_bool'])['whole_weight'].mean()
````

- apply 함수 사용

- DataFrame타입의 객체에서 호출가능한 apply 함수에 대해 알아보자
- 본인이 원하는 행과 열에 연산 혹은 function을 적용할 수 있다.
- 열 기준으로 집계하고 싶은 경우 axis=1
- 행 기준으로 집계하고 싶은 경우 axis=0

````python
df[['diameter']].apply(np.average,axis=0)
````

````python
# 행기준 집계
df[['diameter','whole_weight']].apply(np.average,axis=0)
````

````python
# 열 기준 집계
df[['diameter','whole_weight']].apply(np.average, axis=1)
````

````python
## 사용자 함수를 통한 집계
import math
def avg_ceil(x,y,z):
    return math.ceil((x+y+z)/3)
df[['diameter','height','whole_weight']].apply(lambda x : avg_ceil(x[0],x[1],x[2]),axis=1)
````

````python
from pandas import read_excel
import numpy
df = read_excel("https://data.hossam.kr/C02/pre_sample.xlsx", index_col='이름')
df
````

````python
df5 = df.copy()
df5
````

````python
df5['합계'] = df.loc[:, '국어':'과학'].sum(axis=1)
df5['평균'] = df.loc[:, '국어':'과학'].mean(axis=1)
df5
````

````python
# 평균에 따른 조건부 파생변수
df5['결과'] = numpy.where(df5['평균'] >=80, '합격', '불합격')
# df5
````

````python
# 경우에 수에 따른 조건 값
conditions = [ (df5['평균']>=90), #'A',
              (df5['평균']>=80), #'B',
              (df5['평균']>=70), #'C' 
             ]
value = ["A", "B", "C"]
df5['학점'] = numpy.select(conditions, value, default='F')
df5
````
