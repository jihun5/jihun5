머신러닝은 레이블(label, 정답)이 있는 입력(input) 데이터를 모델에 투입하여 머신러닝 모델이 입력 데이터와 정답 사이의 관계를 갖는다.
모델이 파악한 관계식을 적용하여 결과를 예측할 수 있거나 패턴이나 규칙을 알아내는 것도 가능하다.

머신러닝은 정답레이블을 알고 있는 지 여부에 따라 두가지 유형으로 나뉜다.

지도학습 : 입력데이터 X와 출력데이터 Y를 모두 알고 있는 상태에서, Y=aX +b와 같이 X와 Y 사이의 관계식을 알아내는 머신러닝 알고리즘을 말한다.
이때 에측 목표가 되는 Y변수를 목표변수(target)라고도 하고 목표변수를 예측하는데 사용되는 X변수를 설명변수(Feature)라고 부른다.
즉, 데이터 X가 주어졌을 때, 모델링 관계식에 X를 대입하면 Y를 예측할 수 있다.

비지도 학습은 모델 학습 단계에서 정답 Y가 주어지지 않고 X데이터만 제공되는 머신러닝 유형이다. 따라서 정답 Y를 예측하는 것이 목표가 아닌 X데이터 사이에 존재하는 패턴 또는 규칙을 찾는 것이 목표다. 

회귀문제는 설명 변수(X)와 목표 변수(Y) 사이의 회귀관계식을 갖는다. 목표변수는 연속적인 값을 갖는 숫자형 데이터이다. 
EX- 주가 데이터를 가지고 미래 주가를 예측하거나, 자동차 배기량이나 연식 등으로 중고차 정보를 이용하여 가격을 예측하는 문제를 예로 들 수 있다.

분류문제는 설명 변수(X)와 목표 변수(Y) 사이의 관계를 찾지만, 예측하려는 목표 레이블이 연속적이지 않고 0,1,2와 같이 이산적인 값을 갖는 경우를 말한다. 
EX - 개 또는 고양이를 찍은 사진을 읽어서 개(클래스 0)인지 고양이(클래스 1)인지 분류하는 문제는 이진 분류에 속한다
0~9까지의 숫자를 기록한 숫자 카드를 읽는 경우 어떤 카드 인지 파악하는 문제는 다중 분류에 속한다.

머신러닝 프로세스
경진 대회에 처음 참가하는 상황을 가정하고, 데이터를불러오는 것부터 시작해서 가공 및 모델 학습을 거쳐 결과물을 제출하는 과정을 단계적으로 설명해보자

1.문제 파악 : 경진 대회에서 주어지는 문제를 파악한다. 해당 분야의 도메인 지식을 습득하는 것지 좋다.

2. 데이터 탐색: 대회에서 제공하는 데이터를 내부 데이터라고 하면 외부 데이터는 공공데이터 포털 등 외부에서 수집하는 데이터를 말한다. 데이터에서 허용하는 범위 내에서 외부 데이터를 읽고 누락, 이상치를 확인한다. 그리고 데이터 구조 및 특성을 파악하고 데이터의 분포와 상관관계를 밝힌다.

3. 데이터 전처리: 데이터 탐색이 끝나면 모델 학습이 가능한 형태로 데이터를 정리한다, 내부데이터와 외부데이터를 병합하고, 데이터 탐색 단계에서 확인할 결과값과 이상치를 처리한다. 필요한 경우 자료형을 변환한다. *자주나오는 실수는 문자열 데이터나 이미지형 데이터의 경우 숫자형으로 변환하는 작업이 필요하다. 새로운 속성을 추가하거나, 불필요한 속성을 제거하기도한다.


4.모델 학습 : 모델 학습에 필요한 훈련 데이터와 모델 성능을 평가하기 위한 검증 데이터로 구분한다. 예측 알고리즘을 선택하고 모델을 설계한다. 훈련데이터를 입력하여 모델을 학습시키고 검증 데이터를 입력하여 학습을 마친 모델의 성능을 평가한다.
모델 성능을 높일 수 있도록 하이퍼파라미터를 튜닝하고 최종 모델을 선택한다

5. 예측 : 예측해야 하는 테스트 데이터를 모델에 입력하고, 모델이 예측한 값을 제출용 파일의 형식에 맞게 정리한다. 모델의 일반화 성능을 확인하고 리더보드 점수를 올릴 수 있는 방향으로 모델을 수정하고 제출하는 과정을 반복한다.
